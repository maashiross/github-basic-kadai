```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} さんは大人です。")
        else:
            print(f"{self.name} さんは大人ではありません。")
humans = [
    Human("侍太郎", 25),
    Human("侍花子", 18),
    Human("侍二郎", 20),
    Human("侍三郎", 15)
]
for person in humans:
    person.check_adult()
```
